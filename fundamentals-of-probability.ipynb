# 1. Introduction to Probability

Probability is a branch of mathematics that deals with calculating the likelihood of a given event's occurrence. This measure of likelihood is quantified as a number between 0 and 1. A probability of 0 means the event will not happen, and a probability of 1 means the event will certainly happen.

# Definition of Probability

# Python Code

**Calculating Probability**

```python
# Example: Probability of rolling an even number with a fair six-sided die

# Sample space (S)
sample_space = {1, 2, 3, 4, 5, 6}

# Event (E): Rolling an even number
event_e = {2, 4, 6}

# Calculate probability P(E)
probability_e = len(event_e) / len(sample_space)
print("Probability of rolling an even number:", probability_e)
```

# Python Code 

**Conditional Probability**

```python
# Example: Conditional probability of drawing a red card from a deck given it's a face card

# Sample space (S): Deck of cards
deck = {'red': {'heart', 'diamond'}, 'black': {'spade', 'club'}}

# Event (A): Drawing a face card
event_a = {'heart', 'diamond', 'spade', 'club'}

# Event (B): Drawing a red card
event_b = {'heart', 'diamond'}

# Calculate conditional probability P(A|B)
probability_a_given_b = len(event_a & event_b) / len(event_b)
print("Conditional Probability of drawing a face card given a red card:", probability_a_given_b)
```

# 4. Random Variables

# Definition and Types: Discrete and Continuous
A random variable is a variable whose value is subject to variations due to chance. It can take on different values, each with an associated probability.

- **Discrete Random Variable:** Takes on a countable number of distinct values. Examples include the number of heads in a series of coin tosses or the number of students in a classroom.

- **Continuous Random Variable:** Takes on an infinite number of possible values within a given range. Examples include the height of students in a classroom or the time required to run a marathon.

# Python Code

**1. Discrete Random Variable**

```python
import numpy as np

# Parameters for Poisson distribution
lam = 3  # Average rate

# Generate random samples from Poisson distribution
samples = np.random.poisson(lam, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Poisson random variable: {mean}")
print(f"Variance of the Poisson random variable: {variance}")
```

**2. Continuous Random Variable**

```python
import numpy as np

# Parameters for Normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate random samples from Normal distribution
samples = np.random.normal(mu, sigma, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Normal random variable: {mean}")
print(f"Variance of the Normal random variable: {variance}")
```

# 5. Probability Distributions: Discrete Distributions

# Python Code 

**Binomial distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for Binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate random samples from Binomial distribution
samples = np.random.binomial(n, p, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Binomial random variable: {mean}")
print(f"Variance of the Binomial random variable: {variance}")

# Plot PMF (Probability Mass Function)
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.show()
```

# Python Code 

**Poisson distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter for Poisson distribution
lam = 3  # Average rate

# Generate random samples from Poisson distribution
samples = np.random.poisson(lam, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Poisson random variable: {mean}")
print(f"Variance of the Poisson random variable: {variance}")

# Plot PMF (Probability Mass Function)
x = np.arange(0, 10)
pmf = poisson.pmf(x, lam)

plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Poisson Distribution PMF')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```

# Python Code

**Geometric Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parameter for Geometric distribution
p = 0.3  # Probability of success

# Generate random samples from Geometric distribution
samples = np.random.geometric(p, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Geometric random variable: {mean}")
print(f"Variance of the Geometric random variable: {variance}")

# Plot PMF (Probability Mass Function)
x = np.arange(1, max(samples) + 1)
pmf = geom.pmf(x, p)

plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Geometric Distribution PMF')
plt.xlabel('Number of trials until first success')
plt.ylabel('Probability')
plt.show()
```

# Python Code

**Negetive Binomial Distribuation**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters for Negative Binomial distribution
r = 5  # Number of successes
p = 0.3  # Probability of success

# Generate random samples from Negative Binomial distribution
samples = np.random.negative_binomial(r, p, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Negative Binomial random variable: {mean}")
print(f"Variance of the Negative Binomial random variable: {variance}")

# Plot PMF (Probability Mass Function)
x = np.arange(r, np.max(samples) + 1)
pmf = nbinom.pmf(x, r, p)

plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Negative Binomial Distribution PMF')
plt.xlabel('Number of trials until r successes')
plt.ylabel('Probability')
plt.show()
```

# Python Code 

**Hyper-geometric Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parameters for Hypergeometric distribution
N = 50   # Population size
K = 20   # Number of successes in the population
n = 10   # Number of draws (samples)

# Generate random samples from Hypergeometric distribution
samples = np.random.hypergeometric(K, N - K, n, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Hypergeometric random variable: {mean}")
print(f"Variance of the Hypergeometric random variable: {variance}")

# Plot PMF (Probability Mass Function)
x = np.arange(max(0, n - (N - K)), min(n, K) + 1)
pmf = hypergeom.pmf(x, N, K, n)

plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Hypergeometric Distribution PMF')
plt.xlabel('Number of successes in sample')
plt.ylabel('Probability')
plt.show()
```

# Python Code

**Gamma Distribuation**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for Gamma distribution
k = 2.0  # Shape parameter (alpha)
theta = 2.0  # Scale parameter (beta)

# Generate random samples from Gamma distribution
samples = np.random.gamma(k, theta, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Gamma random variable: {mean}")
print(f"Variance of the Gamma random variable: {variance}")

# Plot PDF (Probability Density Function)
x = np.linspace(0, 20, 100)
pdf = gamma.pdf(x, k, scale=theta)

plt.plot(x, pdf, label=f'Gamma PDF (k={k}, theta={theta})')
plt.title('Gamma Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

**t-Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parameters for t-distribution
df = 5  # Degrees of freedom

# Generate random samples from t-distribution
samples = np.random.standard_t(df, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the t-distributed random variable: {mean}")
print(f"Variance of the t-distributed random variable: {variance}")

# Plot PDF (Probability Density Function)
x = np.linspace(-5, 5, 100)
pdf = t.pdf(x, df)

plt.plot(x, pdf, label=f't-distribution PDF (df={df})')
plt.title('t-Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

**Logistic distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import logistic

# Parameters for Logistic distribution
loc = 0.0   # Location parameter (mean)
scale = 1.0  # Scale parameter (standard deviation)

# Generate random samples from Logistic distribution
samples = np.random.logistic(loc, scale, 1000)

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Logistic random variable: {mean}")
print(f"Variance of the Logistic random variable: {variance}")

# Plot PDF (Probability Density Function)
x = np.linspace(-10, 10, 100)
pdf = logistic.pdf(x, loc, scale)

plt.plot(x, pdf, label=f'Logistic PDF (loc={loc}, scale={scale})')
plt.title('Logistic Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

# Python Code

**Cauchy distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

# Parameters for Cauchy distribution
x0 = 0.0   # Location parameter
gamma = 1.0  # Scale parameter

# Generate random samples from Cauchy distribution
samples = np.random.standard_cauchy(1000) * gamma + x0

# Compute mean and variance (note: Cauchy distribution has undefined mean and variance)
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Cauchy random variable: {mean} (Note: Cauchy distribution has undefined mean)")
print(f"Variance of the Cauchy random variable: {variance} (Note: Cauchy distribution has undefined variance)")

# Plot PDF (Probability Density Function)
x = np.linspace(-10, 10, 100)
pdf = cauchy.pdf(x, x0, gamma)

plt.plot(x, pdf, label=f'Cauchy PDF (x0={x0}, gamma={gamma})')
plt.title('Cauchy Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

# Python Code

**Weibull distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Parameters for Weibull distribution
k = 1.5   # Shape parameter (k > 0)
lambda_ = 2.0  # Scale parameter (lambda > 0)

# Generate random samples from Weibull distribution
samples = np.random.weibull(k, 1000) * lambda_

# Compute mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of the Weibull random variable: {mean}")
print(f"Variance of the Weibull random variable: {variance}")

# Plot PDF (Probability Density Function)
x = np.linspace(0, 5, 100)
pdf = weibull_min.pdf(x, k, scale=lambda_)

plt.plot(x, pdf, label=f'Weibull PDF (k={k}, lambda={lambda_})')
plt.title('Weibull Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
