# Introduction
Linear algebra is foundational for understanding and implementing deep learning algorithms. This notebook provides a comprehensive exploration of essential concepts with practical Python implementations using NumPy.



# 1. Scalars, Vectors, and Matrices

### Scalars, vectors, and matrices are fundamental building blocks in linear algebra:

#### **Scalars**: Single numerical values.
#### **Vectors**: Arrays of scalars arranged in a specific order.
#### **Matrices**: 2-dimensional arrays of scalars.

# Python Code

Let's start by defining scalars, vectors, and matrices using NumPy and perform basic operations.


```python
import numpy as np

# Scalars
scalar = 5

# Vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Operations
print("Vector Addition:", vector1 + vector2)
print("Matrix Multiplication:\n", np.matmul(matrix1, matrix2))
```

# 2. Matrix Operations

## Matrix operations include addition, subtraction, and multiplication:

### **Matrix Addition:** Element-wise addition of matrices.
### **Matrix Multiplication:** Dot product of matrices.

# Python Code

Implementing matrix operations using NumPy:

```python
# Matrix multiplication
result_matrix = np.matmul(matrix1, matrix2)
print("Matrix Multiplication Result:\n", result_matrix)
```

# 3. Linear Transformations

Linear transformations use matrices to transform vectors in space:

#### **Transformation Matrix:** Matrix that applies transformations like scaling or rotation.


# Python Code

Visualizing linear transformations with sample data:

```python
import matplotlib.pyplot as plt

# Sample data
data = np.random.rand(2, 100)  # 2D data points

# Transformation matrix
transformation_matrix = np.array([[2, 0.5], [0.5, 2]])

# Apply transformation
transformed_data = np.matmul(transformation_matrix, data)

# Plot original and transformed data
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(data[0], data[1])
plt.title('Original Data')
plt.subplot(122)
plt.scatter(transformed_data[0], transformed_data[1])
plt.title('Transformed Data')
plt.show()
```

# 4. Norms and Dot Products

Norms measure the size of vectors, while dot products measure their similarity:

#### **Norm**: Euclidean distance or magnitude of a vector.
#### **Dot Product:** Scalar product of vectors.

# Python Code
Calculating norms and dot products using NumPy:

```python
# Norm of a vector
vector = np.array([3, 4])
norm = np.linalg.norm(vector)
print("Norm of Vector:", norm)

# Dot product of vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
dot_product = np.dot(vector1, vector2)
print("Dot Product:", dot_product)
```

# 5. Eigenvalues and Eigenvectors

Eigenvalues and eigenvectors are critical in linear transformations and matrix diagonalization:

#### **Eigenvalues:** Scalars that scale eigenvectors.
#### **Eigenvectors:** Vectors that remain unchanged in direction during linear transformation.


# Python Code
Computing eigenvalues and eigenvectors using NumPy:

```python
# Eigen decomposition
matrix = np.array([[1, 2], [2, 1]])
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

# 6. Singular Value Decomposition (SVD)

SVD decomposes matrices into singular values and orthogonal matrices:

#### **SVD:** Useful for data compression and dimensionality reduction.

# Python Code
Implementing SVD using NumPy:



```python
# SVD
matrix = np.array([[1, 2], [3, 4], [5, 6]])
U, S, VT = np.linalg.svd(matrix)
print("U matrix:\n", U)
print("S matrix:\n", np.diag(S))
print("VT matrix:\n", VT)
```
